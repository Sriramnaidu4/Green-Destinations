```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("greendestination (1).csv")

# Quick look
print(df.head())
print(df.columns)
```

```python

# 1. Overall Attrition Rate
attrition_rate = df['Attrition'].value_counts(normalize=True) * 100
plt.figure(figsize=(6,6))
attrition_rate.plot(kind='pie', autopct='%1.1f%%', colors=['lightgreen','salmon'])
plt.title("Attrition Rate")
plt.ylabel("")
plt.show()

```

```python
# 2. Attrition vs Age
plt.figure(figsize=(8,6))
sns.boxplot(x='Attrition', y='Age', data=df, palette='Set2')
plt.title("Attrition by Age")
plt.show()

```

```python

# 3. Attrition vs Years at Company
plt.figure(figsize=(8,6))
sns.histplot(data=df, x='YearsAtCompany', hue='Attrition', multiple='stack', palette='Set1')
plt.title("Attrition by Years at Company")
plt.show()

```

```python

# 4. Attrition vs Income
plt.figure(figsize=(8,6))
sns.boxplot(x='Attrition', y='MonthlyIncome', data=df, palette='Set3')
plt.title("Attrition by Monthly Income")
plt.show()

```

```python

# 5. Correlation Heatmap (Extra Insight)
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
